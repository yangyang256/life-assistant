```python
import torch

# 1. æ¨¡æ‹Ÿåœºæ™¯ï¼š
# å‡è®¾æˆ‘ä»¬æœ‰ä¸¤ä¸ªå¼ é‡ï¼ˆå¯ä»¥ç†è§£ä¸ºé‡å­çº¿è·¯ä¸­çš„ä¸¤ä¸ªé—¨ï¼Œæˆ–è€…æ˜¯ç¥ç»ç½‘ç»œåˆ†è§£åçš„ä¸¤ä¸ªæ ¸ï¼‰
# Tensor A: å½¢çŠ¶ [10, 20, 5] -> æŒ‡æ ‡ä¸º (i, j, k)
# Tensor B: å½¢çŠ¶ [5, 20, 30] -> æŒ‡æ ‡ä¸º (k, j, l)
# æ³¨æ„ï¼šç»´åº¦ '20' (j) å’Œ '5' (k) æ˜¯æˆ‘ä»¬è¦æ”¶ç¼©ï¼ˆæ±‚å’Œï¼‰æ‰çš„â€œè…¿â€ã€‚

# éšæœºåˆå§‹åŒ–å¼ é‡ (Batch size=10, Input_dim=20, Rank=5)
tensor_A = torch.randn(10, 20, 5) 
tensor_B = torch.randn(5, 20, 30)

print(f"Tensor A shape: {tensor_A.shape}")
print(f"Tensor B shape: {tensor_B.shape}")

# 2. ä¼ ç»Ÿå†™æ³• (éå¸¸ç—›è‹¦ï¼Œå®¹æ˜“å†™é”™ç»´åº¦)
# é€šå¸¸éœ€è¦ permute æ¢è½´ + reshape + matmulï¼Œæå…¶éš¾è¯»
# æ¯”å¦‚: torch.matmul(tensor_A.permute(...), ...) -> æå…¶åäººç±»

# 3. é‡å­å¯å‘å¼å†™æ³• (Einsum)
# è¯­æ³•å«ä¹‰ï¼š"ijk, kjl -> il"
# i, j, k æ˜¯ A çš„è½´
# k, j, l æ˜¯ B çš„è½´
# -> il è¡¨ç¤ºï¼šä¿ç•™ i å’Œ lï¼ŒæŠŠå‡ºç°ä¸¤æ¬¡çš„ j å’Œ k è‡ªåŠ¨æ±‚å’Œï¼ˆæ”¶ç¼©ï¼‰æ‰ï¼
result = torch.einsum('ijk, kjl -> il', tensor_A, tensor_B)

print("-" * 30)
print(f"Result shape: {result.shape}") 
# ç»“æœåº”è¯¥æ˜¯ [10, 30]
# ç‰©ç†æ„ä¹‰ï¼šæˆ‘ä»¬æŠŠä¸­é—´å¤æ‚çš„çº ç¼ /è¿æ¥(j, k)éƒ½ç®—å®Œäº†ï¼Œåªç•™ä¸‹äº†ä¸¤å¤´çš„è‡ªç”±ç´¢å¼•(i, l)ã€‚

# 4. éªŒè¯ä¸€ä¸‹è®¡ç®—æ˜¯å¦æˆåŠŸ
print("\nğŸ‰ æ­å–œï¼ä½ åˆšåˆšå®Œæˆäº†ä¸€æ¬¡å¼ é‡ç½‘ç»œæ”¶ç¼©ã€‚")
print("åœ¨é‡å­å¯å‘å¼æ¨¡å‹ä¸­ï¼Œè¿™ç›¸å½“äºæŠŠä¸¤ä¸ªåŸæœ¬å‚æ•°é‡å·¨å¤§çš„å±‚ï¼Œé€šè¿‡ä½ç§©åˆ†è§£å†é‡ç»„äº†ã€‚")
```
